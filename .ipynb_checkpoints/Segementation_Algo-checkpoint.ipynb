```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
org_image = cv.imread('C:/Users/pp/Project_DRDO_1/yes/Y30.jpg', cv.IMREAD_GRAYSCALE)


# Define the radius and create the disk-shaped structuring element
radius = 25
size = 2*radius + 1

structure_element = cv.getStructuringElement(cv.MORPH_ELLIPSE, (size, size))

# Perform morphological opening
opened_image = cv.morphologyEx(org_image, cv.MORPH_OPEN, structure_element)

opened_image.astype('uint8')
org_image.astype('uint8')


# Subtract the opened image from the original image to get the skull image
skull_image = cv.subtract(org_image, opened_image)


# Subtract the skull image from the opened image to get the clear skull image
clear_skull_image = cv.subtract(skull_image, opened_image)


output_image = cv.subtract(org_image, clear_skull_image)


#Complete



unique_values = np.unique(output_image)
unique_values = unique_values[unique_values != 0]
n = len(unique_values)
sum_of_unique_values = np.sum(unique_values)
threshold_value = sum_of_unique_values / n

# Binarize the image using the calculated threshold value
binary_image = (org_image > threshold_value).astype(np.uint8) * 255

# Apply a 5x5 median filter to the binary image
median_filtered_image = cv.medianBlur(binary_image, ksize=5)




# Display the results
plt.figure(figsize=(12, 6))


plt.subplot(231), plt.imshow(opened_image, cmap='gray')

plt.subplot(232), plt.imshow(skull_image, cmap='gray'), plt.title('Skull Image')
plt.subplot(233), plt.imshow(clear_skull_image, cmap='gray'), plt.title('Clear Skull Image')
plt.subplot(234), plt.imshow(output_image, cmap='gray'), plt.title('Output Image')
plt.subplot(235), plt.imshow(binary_image, cmap='gray'), plt.title('Binary Image (Thresholded)')
plt.subplot(236), plt.imshow(median_filtered_image, cmap='gray'), plt.title('Median Filtered Image')


plt.tight_layout()
plt.show()
```
